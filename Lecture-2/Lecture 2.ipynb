##Exercise A1: Download the list of characters

####Read about the Categorymembers API, which allows you to download all members of a Wikipedia category. What is a category on Wikipedia? 

A category is a collection or grouping of wikipedia pages under a subject. They are written in markup as: [[Category:XYZ]]

####Explain in a few words the usage and role of the main parameters. Which parameters will you need to download the full list?

* action = query : specifies we want to receive data
* cmtitle = Category:Marvel_Comics_superheroes : specifies which category we want to get
* list = categorymembers : specifies we want the category members to be listed
* cmlimit = 500 : specifies the number of page we want to see (the limit is 500)
* format = json : specifies the format we want (json)

As there are over 500 superheroes and the limit of pages we can see at a time is 500, we need to make two requests, the second request using the cmcontinue parameter from the first, which specifies we want to fetch the next part of our initial query.

####Use the API to obtain the list of superheroes. Print the first 10 results

```python
import urllib2
import json
import pprint
url = "http://en.wikipedia.org/w/api.php?action=query&list=categorymembers&cmtitle=Category:Marvel_Comics_superheroes&cmlimit=1000&format=json"
response = urllib2.urlopen(url)
json_response = json.load(response)
superheroes = json_response["query"]["categorymembers"]
# gets the next parts of the query
while("query-continue" in json_response):
    url += "&cmcontinue="+json_response["query-continue"]["categorymembers"]["cmcontinue"]
    response = urllib2.urlopen(url)
    json_response = json.load(response)
superheroes += json_response["query"]["categorymembers"]
pprint.pprint(superheroes[:10])
```

```python
print(len(superheroes))
```

We see there are now 923 superheroes

##Exercise A2: Download each character’s wiki page and save to your own computer

####Use your experience from last time. For each of the super heroes obtained in the previous exercise, use the Query API to download the full page content and save it to a text file (use the json format)

```python
import urllib2
import json
import urllib

url = "http://en.wikipedia.org/w/api.php?&action=query&titles={}&prop=revisions&rvprop=content&format=json"
superhero_dictionary = {}
for page_title in superheroes_page_titles:
    page_url = url.format(urllib.quote(page_title))
    response = urllib2.urlopen(page_url)
    json_response = json.loads(response.read())
    page_id = json_response["query"]["pages"].keys()[0]
    superhero_dictionary[page_title] = json_response["query"]["pages"][page_id]
with open("superheroes.txt","w+") as f:
    json.dump(superhero_dictionary, f)
```

##Exercise A3: Fun with regular expressions

####Now explain in your own words: what are regular expressions?

Regular expressions are strings used to match text or search for patterns. It can be used to search texts or validate input.

Common use-cases are for example defining regular expressions that validates if a string is an email or a street name and more.

####Provide an example of a regex to match 4 digits numbers

we use the \d expression and specifies 4 repetitions as well as a word boundary on each side so it will only catch 4 digits that stand alone and won't catch the "1234" in "12345" or "1234test"

```python
import re
pattern = r"\b\d{4}\b"

valid = ["1234","5678","9123"]
not_valid = ["12345","123","not_a_number","12345test"]
for string in valid:
    print(re.search(pattern,string))
for string in not_valid:
    print(re.search(pattern,string))
```

####Provide an example of a regex to match words starting with super

super specifices it should start with super, word boundaries to only catch words that stand alone, [a-z] matches any lower-case character from a-z and * is one or more reptitions.

```python
import re
pattern = r"\bsuper[a-z]*\b"

valid = ["superman","super","superduperman"]
not_valid = ["suppemand","123","a_text","super1234"]
for string in valid:
    print(re.search(pattern,string))
for string in not_valid:
    print(re.search(pattern,string))
```

####Provide an example of a regex to match words enclosed between << and >>

"\<" and "\>" matches the characters, {2} specifices there should be 2 of them, [a-zA-Z\s] matches any lower- or upper-case characters from a-z and \s matches a whitespace, + specifices there should be one or more.

```python
import re
pattern = r"<{2}[a-zA-Z\s]+>{2}"

valid = ["<<hello>>","<<hello you>>"]
not_valid = ["word","123","<<1234>>","<<>>"]
for string in valid:
    print(re.search(pattern,string))
for string in not_valid:
    print(re.search(pattern,string))
```

####Infobox sections for superheroes start with {{Superherobox and end with }}. Write a regex to match the infobox text

Apparently there is inconsistent naming of superhero infoboxes on Wikipedia, some start with "Superherobox" others with "Infobox comics character" and yet others with just "Infobox comics". I've tried to create a regex that matches them all.

```python
import re

pattern = r"{{Infobox comics[\S\s]*?}}|{{Superherobox[\S\s]*?}}"

valid = ["{{Superherobox info about superhero}}","{{Superherobox info}}",
         "{{Superherobox info }} more info {{}} end of info }}"]
not_valid = ["word","123","{{<<1234>>}}","<<>>"]
for string in valid:
    print(re.search(pattern,string))
for string in not_valid:
    print(re.search(pattern,string))
```

####Plot the length of pages versus length of infoboxes. Comment on the results

```python
page_vs_infoboxes_list = []
pattern = r"{{Infobox comics[\S\s]*?}}|{{Superherobox[\S\s]*?}}"

for hero in superhero_dictionary.values():
    page_length = len(hero["revisions"][0]["*"])
    
    match = re.search(pattern, hero["revisions"][0]["*"])
    if not match:
        infobox_length = 0
    else:
        infobox_length = len(match.group(0))
    page_vs_infoboxes_list.append((page_length, infobox_length))

    plt.scatter([x[0] for x in page_vs_infoboxes_list],
            [x[1] for x in page_vs_infoboxes_list], marker=".")

plt.xlabel("page length")
plt.ylabel("infobox length")
plt.title("page length vs infobox length")
plt.show()
```

The plot shows page length plotted against infobox length.

if a infobox could not be found, an infobox length of 0 is added.

It seems a number of superheroes wikipedia pages does not contain an infobox, and seemingly page length is not a factor of whether a superhero has an infobox or not.

The relationship between page length and infobox seems to be very mildly correlated, but i thought the correlation would be larger.

##Exercise B4: Timeline of character creation

####For each superhero, extract the debut year from the infobox. Do all pages have a debut? Are there heroes with more than one debut year? How did you handle invalid/missing data?

Outputting if possible debut sentence and found year for each superhero:

```python
superhero_years = []
pattern = r"debut[\S\s]*?(\d{4}){1}?[\S\s]*?(\d{4})[\S\s]*?\|"


for hero, info in superhero_dictionary.items():
    match = re.search(pattern, info["revisions"][0]["*"])
    if match:
        year = match.group(1)
        if year == "2099":
            year = match.group(2)
        print("{} has debut year: {}".format(hero,year))
        superhero_years.append((hero, int(year)))
        print("\n")
    else:
        print("{} does not have a debut year".format(hero))
        print("\n")
        superhero_years.append((hero, 0))
```

From the output we can see that not all found superheroes have a debut year, and some have a debut sentence with several years.

We will take the first debut year found in the case of multiple except when the debut year is found as 2099 as that is a special xmen comic in that case we will use the second.

If a hero does not have a debut year, we will look at them as a group.

Visualising number of superhero debuts per year by using a histogram:

It seems like a decent amount of superheroes debuted in the late 30's and early 40's when Marvel was just created.

Then not much happened in terms of debuts in the late 40's and 50's, in the 60's however superheroes started to boom and the superhero debuts levels look very much alike throughout the next 5 decades except for a dip in the late 90's.

```python
plt.figure(figsize=(20,6))
range_tuple = (1939, 2030)
plt.hist([x[1] for x in superhero_years if x[1] != 0], bins=range_tuple[1]-range_tuple[0], range=range_tuple)
plt.title("Histogram of superhero debut years")
plt.ylabel("number of superheroes")
plt.xlabel("debut year")
plt.show()
```

##Exercise B5: Superheroes Teams

####Teams are saved in the alliances field in the infobox. Write a regex to match the values of the alliances field

```python
superhero_alliances = []
pattern = r"alliances.*?=.*?\[\[([\S\s]*?)\]\](\<br[\s\S\/]\>)?[\s\S]*?\|{1}?[\s\S]*?=[\s\S]*?[}]?"
deep_pattern = r"[\[]{0,2}\|?([\w\s\.\-\']+)[\]]{2}"

for hero, info in superhero_dictionary.items():
    print(hero)
    print("------------------------")
    match = re.search(pattern, info["revisions"][0]["*"])
    if match:
        match = match.group(0)
        alliances = re.findall(deep_pattern, match)
        for alliance in alliances:
            print(alliance)
        superhero_alliances.append(alliance.strip())
    else:
        print("{} has no alliances or infobox".format(hero))
    print("\n")
```

####Count the number of members for each team. Visualize the results and comment

There are a lot of teams, so i'm trying to visualize them using a horizontal bar chart:

```python
alliances_set = set(superhero_alliances)
num_range = range(len(alliances_set))
plt.figure(figsize=(10, 50))
plt.barh(num_range, [superhero_alliances.count(a) for a in alliances_set])
plt.yticks([n + 0.8/2.0 for n in num_range], [a for a in alliances_set])
plt.title("bar chart of superhero alliances")
plt.ylabel("alliance")
plt.xlabel("number of superheroes affiliated with alliance")
plt.axis("tight")
plt.show()
```

It seems the alliances which contain the most superheroes by a quick glance are Avengers Academy, Jean Grey School of Students, X-Men, Avengers, New Warriors.

The alliance with the most members are X-Men with 23 members.

It seems there are a lot of alliances with only one superhero.
