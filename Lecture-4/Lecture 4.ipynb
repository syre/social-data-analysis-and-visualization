##PART A: download cities location

###Exercise A1: Find the list of top cities in USA

####You will need to parse the table to extract the cities locations

```python
import urllib2
import json
import pprint
url = "http://en.wikipedia.org/w/api.php?action=query&titles=List_of_United_States_cities_by_population&format=json&prop=revisions&rvprop=content"
response = urllib2.urlopen(url)
json_response = json.load(response)
content = json_response["query"]["pages"]
page_content = content[content.keys()[0]]["revisions"][0]["*"]

pprint.pprint(page_content)
```
