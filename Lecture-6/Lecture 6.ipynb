###Exercise B1: Create three networks.

####Load all the pages from superheroes and supervillains that you dowloaded for the previous lectures.

Loading from dictionaries and processing:

```python
import json

def is_page_acceptable(body):
    return not body.startswith("#REDIRECT")

with open("superheroes.txt", "r") as f:
    superhero_dictionary = json.load(f)
with open("supervillains.txt", "r") as f:
    supervillain_dictionary = json.load(f)

supervillain_dictionary =  {key: value["revisions"][0]["*"] for (key, value) in supervillain_dictionary.items() if is_page_acceptable(value["revisions"][0]["*"])}
superhero_dictionary =  {key: value["revisions"][0]["*"] for (key, value) in superhero_dictionary.items() if is_page_acceptable(value["revisions"][0]["*"])}
```

####Build a NetworkX directed graph of the links between pages. For each character page, find all the links to other characters, and for each link add an edge in the graph. Using this strategy, create three distinct networks, one graph for superheroes, one for supervillains, one with all characters

creating superhero and supervillain link dictionaries:

```python
import re
superhero_links = {}
supervillain_links = {}
link_pattern = "\[\[([\w\s.,]*)(\|[\w\s]*)?\]\]"

for hero, page in superhero_dictionary.items():
    link_list = re.findall(link_pattern, page)
    link_list = [l[0] for l in link_list if l[0] in superhero_dictionary or l[0] in supervillain_dictionary]
    superhero_links[hero] = link_list

for villain, page in supervillain_dictionary.items():
    link_list = re.findall(link_pattern, page)
    link_list = [l[0] for l in link_list if l[0] in superhero_dictionary or l[0] in supervillain_dictionary]
    supervillain_links[villain] = link_list
```

Creating directed graph with only superheroes:

```python
superhero_graph = nx.DiGraph()
for hero, links in superhero_links.items():
    if hero not in superhero_graph:
        superhero_graph.add_node(hero)
    for link in links:
        if link not in superhero_graph[hero] and link in superhero_dictionary:
            superhero_graph.add_edge(hero, link)

supervillain_graph = nx.DiGraph()
for villain, links in supervillain_links.items():
    if villain not in supervillain_graph:
        supervillain_graph.add_node(villain)
    for link in links:
        if link not in supervillain_graph[villain] and link in supervillain_dictionary:
            supervillain_graph.add_edge(villain, link)

all_graph = nx.DiGraph()
for member, links in supervillain_links.items()+superhero_links.items():
    if member not in all_graph:
        all_graph.add_node(member)
    for link in links:
        if link not in all_graph[member]:
            all_graph.add_edge(member, link)
```

####Why do you think we want to use a directed graph? Could we have used an undirected graph instead?

Because the relationships are not necessarily reciprocal, an undirected graph could be used, but then a relationship would be ambiguous because it could mean A links to B, B links to A or both

##PART C: PageRank

###Exercise C1: In which we explore PageRank, a powerful machine learning tool for estimating importance in graphs.

#### Read about PageRank in “Programming Collective Intelligence”, Chapter 4, pages 54-56 and 69-73 and briefly explain the main idea behind PageRank in your own words. 

The PageRank algorithm calculates the probability that a user will end up at a web page by randomly clicking links, it is calculated from an importance metric of page X and all other pages importance metrics linking to page X as well as the other pages link count.

####Take a look at the code from the book - Use the code from the book to implement your own version of PageRank.

```python
def calculate_page_rank(pages, iterations=20):
    # initialize every page rank to 1
    pageranks = {member:1 for member in pages}
    for i in range(iterations):
        print("Iteration {}".format(i))
        for node in pages:
            pr = 0.15
            for in_edge in pages.in_edges([node]):
                linking_pr = pageranks[in_edge[0]]
                linking_count = len(pages.out_edges([in_edge[0]]))
                pr+= linking_pr/linking_count
            pageranks[node] = pr
    return pageranks
```
