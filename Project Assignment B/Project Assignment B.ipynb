##Data Retrieval

For getting our movie data we use the Internet Movie Database (IMDB), however, as they do not have an openly accessible API for requesting more detailed data about each movie we use it in conjunction with an API called the Open Movie Database API.

The bottom 1000 movies can be found at [this](http://www.imdb.com/search/title?at=0&groups=bottom_1000&sort=user_rating,asc) IMDB page. The top 1000 movies can be found by just changing the parameter values of the URL.

The criteria for making it on the top and bottom IMDB lists are obviously a sufficiently low and high rating but each movie must also have a vote count of at least 1500.

We believe the vote count criteria serves our purpose as we weed out some of the more unknown movies.

We use a Python library called BeautifulSoup to do the scraping and Requests to fetch the web pages themselves.

As we will use the OMDb API to get the rest of the features we are only interested in the IDs of the movies from the IMDB webpage.

First we fetch the bad movies:

```python
import requests
import bs4
import re
bad_params = {"at":"0", "groups":"bottom_1000", "sort":"user_rating,asc", "view":"simple"}
bad_movie_ids = []
base_url = "http://www.imdb.com/search/title"
id_pattern = r"/title/(t{2}\d*)/"
for start in range(1,1001,100):
    bad_params["start"] = start
    r = requests.get(base_url, params=bad_params)
    soup = bs4.BeautifulSoup(r.text)
    id_results = soup.find("table",{"class":"results"}).findAll("td",{"class":"title"})
    rating_results = soup.find("table",{"class":"results"}).findAll("b")
    for entry, td in enumerate(id_results, start=1):
        link = td.find("a")
        id_match = re.search(id_pattern, link["href"])
        rating = float(rating_results[entry].contents[0])
        movie_id = id_match.group(1)
        bad_movie_ids.append(movie_id)
```

Then we fetch the good movies:

```python
good_params = {"at":"0", "groups":"top_1000", "sort":"user_rating,desc","view":"simple"}
good_movie_ids = []
for start in range(1,1001,100):
    good_params["start"] = start
    r = requests.get(base_url, params=good_params)
    soup = bs4.BeautifulSoup(r.text)
    id_results = soup.find("table",{"class":"results"}).findAll("td",{"class":"title"})
    rating_results = soup.find("table",{"class":"results"}).findAll("b")
    for entry, td in enumerate(id_results, start=1):
        link = td.find("a")
        id_match = re.search(id_pattern, link["href"])
        rating = float(rating_results[entry].contents[0])
        movie_id = id_match.group(1)
        good_movie_ids.append(movie_id)
```

We check how many movies are found:

```python
print(len(good_movie_ids))
print(len(bad_movie_ids))
```

As we see scraping the bottom and top 1000 IMDB pages leaves us with 1000 good movies and 399 bad movies.

Next we use OMDb to gather more information about each movie.

first for bad movies:

```python
omdb_params = {"plot":"full","r":"json"}
omdb_base_url = "http://www.omdbapi.com/"
bad_movies = {}
for bad_id in bad_movie_ids:
    omdb_params["i"] = bad_id
    response = requests.get(omdb_base_url,params=omdb_params)
    bad_movies[bad_id] = response.json()
```

Then for good movies:

```python
omdb_params = {"plot":"full","r":"json"}
omdb_base_url = "http://www.omdbapi.com/"
good_movies = {}
for good_id in good_movie_ids:
    omdb_params["i"] = good_id
    response = requests.get(omdb_base_url,params=omdb_params)
    good_movies[good_id] = response.json()
```

We save the data to the disk so we don't have to rely on the API moving forward:

##Preprocessing:

The data we retrieved from OMDB needs to be preprocessed a bit for our purposes, things like runtime in minutes are not saved as a integer (number) but rather as a string (text), thus we need to convert it to an integer.

The runtime feature is saved in the form "XX min", we need to convert it to an integer for every movie, we do this by cutting off the last 4 characters of each string and converting it:

```python
filterlist = []
for key,val in good_movies.items():
    try:
        val["Runtime"] = int(val["Runtime"][:-4])
    except ValueError:
        print("removing movie: {} because runtime is invalid: {}".format(val["Title"], val["Runtime"]))
        filterlist.append(key)
good_movies = {k:v for k,v in good_movies.items() if k not in filterlist}
filterlist = []
for key,val in bad_movies.items():
    try:
        val["Runtime"] = int(val["Runtime"][:-4])
    except ValueError:
        print("removing movie: {} because runtime is invalid: {}".format(val["Title"], val["Runtime"]))
        filterlist.append(key)
bad_movies = {k:v for k,v in bad_movies.items() if k not in filterlist}
```

As we can see some movies have no runtimes available and thus we remove them.

Furthermore as we need the genres, actors, languages, writers and countries of each movie, and they are saved as comma-separated texts we need to convert them to something more easily usable, thus we convert the texts to lists.

Furthermore sometimes if there is several writers or directors, their respective area is enclosed in parenthesis, we split the text on a start parenthesis to only get their name:

```python
for key, val in good_movies.items():
    splitted_genre = val["Genre"].split(",")
    val["Genre"] = [x.strip() for x in splitted_genre]
    
    splitted_language = val["Language"].split(",")
    val["Language"] = [x.strip() for x in splitted_language]
    
    splitted_actors = val["Actors"].split(",")
    val["Actors"] = [x.strip() for x in splitted_actors]
    
    splitted_countries = val["Country"].split(",")
    val["Country"] = [x.strip() for x in splitted_countries]
    
    splitted_writers = val["Writer"].split(",")
    val["Writer"] = [x.strip().split(" (")[0] for x in splitted_writers]
    
    splitted_directors = val["Director"].split(",")
    val["Director"] = [x.strip().split(" (")[0] for x in splitted_writers]


for key, val in bad_movies.items():
    splitted_genre = val["Genre"].split(",")
    val["Genre"] = [x.strip() for x in splitted_genre]
    
    splitted_language = val["Language"].split(",")
    val["Language"] = [x.strip() for x in splitted_language]
    
    splitted_actors = val["Actors"].split(",")
    val["Actors"] = [x.strip() for x in splitted_actors]
    
    splitted_countries = val["Country"].split(",")
    val["Country"] = [x.strip() for x in splitted_countries]
    
    splitted_writers = val["Writer"].split(",")
    val["Writer"] = [x.strip().split(" (")[0] for x in splitted_writers]
        
    splitted_directors = val["Director"].split(",")
    val["Director"] = [x.strip().split(" (")[0] for x in splitted_writers]
```

## Creating a Dataset of all the movies:

Now we will create a dataset of all the movies with the following features:

Nominal: Genre, Year, Language, Country, Rated, Actors, Writers

Ratio: Runtime

We will filter all other features.

As there can be multiple actors, writers,languages and genres for one movie, we use a technique called one-of-k encoding to make each a feature. We could also create a feature for each and enumerate them, however this would make some machine learning algorithms treat them as continuous or ordered data which they are not.

```python
import numpy as np
from sklearn.feature_extraction import DictVectorizer
import itertools

filterlist = ["Year", "Rated", "Runtime"]

unique_actors = set((itertools.chain(*[x["Actors"] for x in list(good_movies.values())+list(bad_movies.values())])))
unique_languages = set((itertools.chain(*[x["Language"] for x in list(good_movies.values())+list(bad_movies.values())])))
unique_genres = set((itertools.chain(*[x["Genre"] for x in list(good_movies.values())+list(bad_movies.values())])))
unique_countries = set((itertools.chain(*[x["Country"] for x in list(good_movies.values())+list(bad_movies.values())])))
unique_writers = set((itertools.chain(*[x["Writer"] for x in list(good_movies.values())+list(bad_movies.values())])))
unique_directors = set((itertools.chain(*[x["Director"] for x in list(good_movies.values())+list(bad_movies.values())])))
good_movies_filtered = {}
for key, value in good_movies.items():
    good_movies_filtered[key] = {}
    for feature in filterlist:
        good_movies_filtered[key][feature] = good_movies[key][feature]
    for actor in unique_actors:
        if actor in value["Actors"]:
            good_movies_filtered[key][actor] = True
        else:
            good_movies_filtered[key][actor] = False
    for language in unique_languages:
        if language in value["Language"]:
            good_movies_filtered[key][language] = True
        else:
            good_movies_filtered[key][language] = False
    for genre in unique_genres:
        if genre in value["Genre"]:
            good_movies_filtered[key][genre] = True
        else:
            good_movies_filtered[key][genre] = False
    for country in unique_countries:
        if country in value["Country"]:
            good_movies_filtered[key][country] = True
        else:
            good_movies_filtered[key][country] = False
    for writer in unique_writers:
        if writer in value["Writer"]:
            good_movies_filtered[key][writer] = True
        else:
            good_movies_filtered[key][writer] = False       
    for director in unique_directors:
        if director in value["Director"]:
            good_movies_filtered[key][director] = True
        else:
            good_movies_filtered[key][director] = False
        

bad_movies_filtered = {}
for key, value in bad_movies.items():
    bad_movies_filtered[key] = {}
    for feature in filterlist:
        bad_movies_filtered[key][feature] = bad_movies[key][feature]
    for actor in unique_actors:
        if actor in value["Actors"]:
            bad_movies_filtered[key][actor] = True
        else:
            bad_movies_filtered[key][actor] = False
    for language in unique_languages:
        if language in value["Language"]:
            bad_movies_filtered[key][language] = True
        else:
            bad_movies_filtered[key][language] = False
    for genre in unique_genres:
        if genre in value["Genre"]:
            bad_movies_filtered[key][genre] = True
        else:
            bad_movies_filtered[key][genre] = False
    for country in unique_countries:
        if country in value["Country"]:
            bad_movies_filtered[key][country] = True
        else:
            bad_movies_filtered[key][country] = False
    for writer in unique_writers:
        if writer in value["Writer"]:
            bad_movies_filtered[key][writer] = True
        else:
            bad_movies_filtered[key][writer] = False
    for director in unique_directors:
        if director in value["Director"]:
            bad_movies_filtered[key][director] = True
        else:
            bad_movies_filtered[key][director] = False

vec = DictVectorizer()
X = vec.fit_transform(list(good_movies_filtered.values())+list(bad_movies_filtered.values())).toarray()
y = np.array([0 for x in good_movies_filtered]+[1 for x in bad_movies_filtered])
```

## Sentiment Analysis with ANEW Scores

Next we will use Sentiment analysis on a subset of movies using their dialogue in textual form (subtitles)

Sentiment analysis is the analysis of contextual polarity (positive or negative emotions) in text.
We will analyze it using ANEW data which is an acronym for Affective Norms of English Words, the dataset describes nouns with respect to three dimensions of emotion it can possess.
The three emotions are:
* Valence (the pleasantness of a stimulus)
* Arousal (the intensity of emotion provoked by a stimulus)
* Dominance (the degree of control exerted by a stimulus)

We start by creating the ANEW dictionary:

```python
import csv
import pprint
anew_dictionary = {}
with open("Ratings_Warriner_et_al.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter="\n")
    for index, row in enumerate(reader):
        splitted = row[0].split(",")
        # print out header
        if index == 0:
            pprint.pprint(",".join([splitted[1], splitted[2], splitted[5], splitted[8]]))
        else:
            anew_dictionary[splitted[1]] = [float(splitted[2]), float(splitted[5]), float(splitted[8])]
for key, val_list in anew_dictionary.items():
    anew_dictionary[key] = [val-5 for val in val_list]
```

We will analyze the dialogue of 50 bad and 50 good movies, we settled on 50 each since we did not find a suitable API for retrieval and thus had to download them by hand, furthermore finding subtitles for some of the bad movies proved to be difficult.

The subtitles are in SubRip (SRT) format and english.

[Wikipedia](http://en.wikipedia.org/wiki/SubRip) states that The SubRip format dictates each line is in the form: 
1. A numeric counter identifying each sequential subtitle
2. The time that the subtitle should appear on the screen, followed by --> and the time it should disappear
3. Subtitle text itself on one or more lines
4. A blank line containing no text, indicating the end of this subtitle

Next we load in the SRT files:

```python
import os.path
bad_movie_subset = {}
bad_root = "bad_movie_subtitles"
for key,movie in bad_movies.items():
    if os.path.isfile(os.path.join(bad_root,key+".srt")):
        with open(os.path.join(bad_root,key+".srt")) as f:
            # cutting off the counter and time indicator for each line
            splitted = [" ".join(text.split("\n")[2:]) for text in f.read().split("\n\n")]
            bad_movie_subset[key] = " ".join(splitted)

good_movie_subset = {}
good_root = "good_movie_subtitles"
for key,movie in good_movies.items():
    if os.path.isfile(os.path.join(good_root,key+".srt")):
        with open(os.path.join(good_root,key+".srt")) as f:
            # cutting off the counter and time indicator for each line
            splitted = [" ".join(text.split("\n")[2:]) for text in f.read().split("\n\n")]
            good_movie_subset[key] = " ".join(splitted)
```

and we generate X_bow, the Bag of Words representation:

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import tree
import numpy as np
# have to sort the dictionary words when giving them as we else cannot rely on index
vectorizer = CountVectorizer(vocabulary=sorted(anew_dictionary.keys()))
movie_subtitles_collection = [x for x in good_movie_subset.values()]+[x for x in bad_movie_subset.values()]

X_bow = vectorizer.fit_transform(movie_subtitles_collection).toarray()
y_bow = np.array([0 for x in good_movie_subset.values()] + [1 for x in bad_movie_subset.values()])
C_bow = ["Good", "Bad"]
```

We define a function for generating our dataset of ANEW weighted Bag of Words representations:

```python
def calculate_scores(movies, bag_of_words, anew_weights):
    movie_scores = {}
    valence_index = 0
    arousal_index = 1
    dominance_index = 2
    anew_keys = sorted(anew_weights.keys())
    for i, movie in enumerate(movies):
        word_sum = 0
        valence_score = 0
        arousal_score = 0
        dominance_score = 0
        for j, word_count in enumerate(bag_of_words[i]):
            dominance_score += word_count*anew_weights[anew_keys[j]][dominance_index]
            arousal_score += word_count*anew_weights[anew_keys[j]][arousal_index]
            valence_score += word_count*anew_weights[anew_keys[j]][valence_index]
            word_sum += word_count
        movie_scores[movie] = [valence_score/word_sum, arousal_score/word_sum, dominance_score/word_sum]
    return movie_scores

movie_names = [x for x in good_movie_subset.keys()]+[x for x in bad_movie_subset.keys()]
movie_scores = calculate_scores(movie_names, X_bow, anew_dictionary)
X_anew = np.array([movie_scores[x] for x in movie_names])
```

```python
import matplotlib.patches as mpatches

fig = plt.figure(figsize=(10,10))
val_vs_arou_ax = fig.add_subplot(221)
val_vs_arou_ax.set_title("Valence vs Arousal")
val_vs_arou_ax.set_xlabel("Valence")
val_vs_arou_ax.set_ylabel("Arousal")
red_patch = mpatches.Patch(color='red', label='bad')
blue_patch = mpatches.Patch(color="blue", label="good")
plt.legend(handles=[red_patch, blue_patch])
val_vs_dom_ax = fig.add_subplot(222)
val_vs_dom_ax.set_title("Valence vs Dominance")
val_vs_dom_ax.set_xlabel("Valence")
val_vs_dom_ax.set_ylabel("Dominance")
arou_vs_dom_ax = fig.add_subplot(223)
arou_vs_dom_ax.set_title("Arousal vs Dominance")
arou_vs_dom_ax.set_xlabel("Arousal")
arou_vs_dom_ax.set_ylabel("Dominance")

good_valence = [movie_scores[m][0] for m in good_movie_subset.keys()]
good_arousal = [movie_scores[m][1] for m in good_movie_subset.keys()]
good_dominance = [movie_scores[m][2] for m in good_movie_subset.keys()]

bad_valence = [movie_scores[m][0] for m in bad_movie_subset.keys()]
bad_arousal = [movie_scores[m][1] for m in bad_movie_subset.keys()]
bad_dominance = [movie_scores[m][2] for m in bad_movie_subset.keys()]

val_vs_arou_ax.scatter(good_valence, good_arousal, c="blue", marker="x", label="good movies")
val_vs_dom_ax.scatter(good_valence, good_dominance, c="blue", marker="x", label="good movies")
arou_vs_dom_ax.scatter(good_arousal, good_dominance, c="blue", marker="x", label="good movies")

val_vs_arou_ax.scatter(bad_valence, bad_arousal, c="red", marker="x", label="bad movies")
val_vs_dom_ax.scatter(bad_valence, bad_dominance, c="red", marker="x", label="bad movies")
arou_vs_dom_ax.scatter(bad_arousal, bad_dominance, c="red", marker="x", label="bad movies")

val_vs_arou_ax.legend()
val_vs_dom_ax.legend()
arou_vs_dom_ax.legend()
plt.show()
```

Valence and dominance seem to be correlated and arousal and dominance seem to be negatively correlated.

By looking at the plots no patterns in terms of significant differences seem to emerge.
