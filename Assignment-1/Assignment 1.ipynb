####Download all Marvel super-hero pages from wikipedia. Save to your hard drive.

First we use the Wikipedia API to obtain the list of superheroes:

```python
import urllib2
import json
import pprint
url = "http://en.wikipedia.org/w/api.php?action=query&list=categorymembers&cmtitle=Category:Marvel_Comics_superheroes&cmlimit=1000&format=json"
response = urllib2.urlopen(url)
json_response = json.load(response)
superheroes = json_response["query"]["categorymembers"]
# gets the next parts of the query
while("query-continue" in json_response):
    url += "&cmcontinue="+json_response["query-continue"]["categorymembers"]["cmcontinue"]
    response = urllib2.urlopen(url)
    json_response = json.load(response)
superheroes += json_response["query"]["categorymembers"]
```

Next we get titles and filters categories and prints 10 first:

```python
superheroes_page_titles = [page["title"] for page in superheroes if not page["title"].startswith("Category:")]
pprint.pprint(superheroes_page_titles[:10])
```

Saving to a dictionary on the disk:

```python
import urllib2
import json
import urllib

url = "http://en.wikipedia.org/w/api.php?&action=query&titles={}&prop=revisions&rvprop=content&format=json"
superhero_dictionary = {}
for page_title in superheroes_page_titles:
    page_url = url.format(urllib.quote(page_title))
    response = urllib2.urlopen(page_url)
    json_response = json.loads(response.read())
    page_id = json_response["query"]["pages"].keys()[0]
    superhero_dictionary[page_title] = json_response["query"]["pages"][page_id]
with open("superheroes.txt","w+") as f:
    json.dump(superhero_dictionary, f)
```

####Use the Query API to obtain the content of the page

```python
import urllib2
import json
import pprint
url = "http://en.wikipedia.org/w/api.php?action=query&titles=List_of_United_States_cities_by_population&format=json&prop=revisions&rvprop=content"
response = urllib2.urlopen(url)
json_response = json.load(response)
content = json_response["query"]["pages"]
page_content = content[content.keys()[0]]["revisions"][0]["*"]
```

####Extract the cities locations.

We extract each row and write regexes for location and pagename and store in a list where each city is a tuple:

```python
import re

cities = []

pattern = "\|(\d[\d\D]*?)\|-\\n"
matches = re.findall(pattern, page_content)
# slicing the end to cut off at the last city
matches = matches[:295]
for match in matches:
    pagename_pattern = "\[\[([\w\s.,]*)(\|[\w\s]*)?\]\]"
    location_pattern = "\|(\d*.\d*)\\xb0N\s*(\d*.\d*)\\xb0W"
    
    pagename_match = re.search(pagename_pattern, match)
    if pagename_match:
        pagename = pagename_match.group(1)
    
    location_match = re.search(location_pattern, match)
    if location_match:
        longitude = float(location_match.group(1))
        # making latitude negative
        latitude = -float(location_match.group(2))
    cities.append((pagename, longitude, latitude))
```

###Viz 2. Visualize the timeline of hero origins. We now investigate whether there is a time trend in the number of superheroes introduced over the years.

####For each superhero, extract the debut year from the infobox.

```python
import re

pattern = r"debut[\S\s]*?(\d{4}){1}?[\S\s]*?(\d{4})[\S\s]*?\|"

def get_super_years(hero_list):
    super_years = []
    for hero, info in hero_list:
        match = re.search(pattern, info)
        if match:
            year = match.group(1)
            if year == "2099":
                year = match.group(2)
            super_years.append((hero, int(year)))
        else:
            super_years.append((hero, 0))
    return super_years
superhero_years = get_super_years(good_heroes)
```

From the output we can see that not all found superheroes have a debut year, and some have a debut sentence with several years.

We will take the first debut year found in the case of multiple except when the debut year is found as 2099 as that is a special xmen comic in that case we will use the second.

If a hero does not have a debut year, we will look at them as a group.

Visualising number of superhero debuts per year by using a histogram:

```python
plt.figure(figsize=(20,6))
range_tuple = (1939, 2030)
plt.hist([x[1] for x in superhero_years if x[1] != 0], bins=range_tuple[1]-range_tuple[0], range=range_tuple)
plt.title("Histogram of Marvel superhero debut years")
plt.ylabel("number of superheroes")
plt.xlabel("debut year")
plt.show()
```

It seems like a decent amount of superheroes debuted in the late 30's and early 40's when Marvel was just created.
Then not much happened in terms of debuts in the late 40's and 50's, in the 60's however superheroes started to boom and the superhero debuts levels look very much alike throughout the next 5 decades except for a dip in the late 90's.

###Viz 3. Superheroes Teams. In this exercise we will extract the teams that each superhero belong to. Teams are saved in the alliances field in the infobox.

####Write a regexp to match the values of the alliances field

Getting alliances for superheroes:

```python
import re
def get_super_alliances(hero_list):
    alliances_list = []
    pattern = r"alliances.*?=.*?\[\[([\S\s]*?)\]\](\<br[\s\S\/]\>)?[\s\S]*?\|{1}?[\s\S]*?=[\s\S]*?[}]?"
    deep_pattern = r"[\[]{0,2}\|?([\w\s\.\-\']+)[\]]{2}"

    for hero, info in good_heroes:
        match = re.search(pattern, info)
        if match:
            match = match.group(0)
            alliances = re.findall(deep_pattern, match)
            for alliance in alliances:
                alliances_list.append(alliance.strip())
    return alliances_list
superhero_alliances = get_super_alliances(good_heroes)
print(superhero_alliances[:10])
```

####Count the number of members for each team. Visualize the results and comment

There are a lot of alliances, so i'll visualize them using a scrollable horizontal bar chart:

```python
import matplotlib.pyplot as plt

alliances_set = set(superhero_alliances)
num_range = range(len(alliances_set))
plt.figure(figsize=(10, 50))
plt.barh(num_range, [superhero_alliances.count(a) for a in alliances_set])
plt.yticks([n + 0.8/2.0 for n in num_range], [a for a in alliances_set])
plt.title("bar chart of Marvel superhero alliances")
plt.ylabel("alliance")
plt.xlabel("number of superheroes affiliated with alliance")
plt.axis("tight")
plt.show()
```

####Why do you think we recommend scaling by the square root?

Because of the high variance of the counts, with the highest city (New York) having about 90 counts and the lowest having 1, New York would be 90 times bigger than the one with only 1 count. This would hurt the visualization by covering the other cities.

By squaring them, we make New York only ~9 times bigger with $sqrt(1)=1$ and $sqrt(90)=9.48$
