####Describe the process of building the tree from training data.   


The algorithm of building the tree is continuously dividing the data set into two groups split based on a criteria/variable.

####Describe how to choose which variable to split on.

You want to split on the variable which creates the least possible mixing, so for example if splitting on a variable color = blue would result in a branch where results are only blueberries, that would be a good splitting variable because no mixing occurs in that branch.

####Describe Gini impurity and entropy.

Gini impurity measures error rate of the set if a result from the set is randomly applied to an item in the set.

For example lets say we have a set of: { Lemon, Lemon} as every item is a Lemon the error rate is 0 as we cannot choose an item from the set that would result in a wrong classification.

If we then have a new set of: {Lemon, Lemon, Orange} the error rate would be $$\frac{num_{lemons}}{total} * \frac{num_{oranges}}{total} + \frac{num_{oranges}}{total} * \frac{num_{lemons}}{total} = 0.444$$

Entropy measures how different the outcomes are to each other. An entropy of 0 means all of the outcomes are the same.

If we have the same set of: {Lemon, Lemon, Orange} the entropy would be $$-\frac{num_{lemons}}{total}*(log(\frac{num_{lemons}}{total})/\log(2)) - \frac{num_{oranges}}{total}*(log(\frac{num_{oranges}}{total})/\log(2)) = 0.918$$

####Describe the recursive tree building process.

The algorithm recursively builds the tree by calculating the information gain for all attributes which is the difference between the current entropy and the average entropy of the two possible new groups.

If the information gain is high and thus the average entropy of the possible two new groups is lower than the current entropy, new branches are created.

These steps are repeated until the information gain is zero.

####What is overfitting? How can pruning of the tree as described in the book cure overfitting?


Overfitting is when a classifier matches the training data too much and thus is more accurate on predicting known data than unknown data. An overfitting classifier fails to generalize when predicting new data. 

Pruning is merging nodes which do not meet a certain criteria like a minimum gain, when the tree has been made.
By pruning the decision tree is allowed to generalize and stops making predictions that might only fit the training data.

####What are missing values? How does the decision tree deal with missing values?

In case of a missing value necessary for following a branch, both branches will be followed but results will be weighted based on the probability that an observation belongs in either category.

####Generate an X_bow and an X_ANEW matrix plus their corresponding header vectors

```python
import json

with open("superheroes.txt", "r") as f:
    superhero_dictionary = json.load(f)
with open("supervillains.txt", "r") as f:
    supervillain_dictionary = json.load(f)
```

creating lists of good, bad and neutral heroes and filters out the REDIRECT pages:

```python
good_heroes = [(key,value["revisions"][0]["*"]) for key,value in superhero_dictionary.items() if key not in supervillain_dictionary]
bad_heroes = [(key,value["revisions"][0]["*"]) for key, value in supervillain_dictionary.items() if key not in superhero_dictionary]
neutral_heroes = [(key,value["revisions"][0]["*"]) for key, value in superhero_dictionary.items() if key in supervillain_dictionary]

def is_page_acceptable(body):
    return not body.startswith("#REDIRECT")

good_heroes = [hero for hero in good_heroes if is_page_acceptable(hero[1])]
bad_heroes = [hero for hero in bad_heroes if is_page_acceptable(hero[1])]
neutral_heroes = [hero for hero in neutral_heroes if is_page_acceptable(hero[1])]
```

creating the dictionary with anew words:

```python
import csv
import pprint
anew_dictionary = {}
with open("Ratings_Warriner_et_al.csv", "rb") as csvfile:
    reader = csv.reader(csvfile, delimiter="\n")
    for index, row in enumerate(reader):
        splitted = row[0].split(",")
        # print out header
        if index == 0:
            pprint.pprint(",".join([splitted[1], splitted[2], splitted[5], splitted[8]]))
        else:
            anew_dictionary[splitted[1]] = [float(splitted[2]), float(splitted[5]), float(splitted[8])]
for key, val_list in anew_dictionary.items():
    anew_dictionary[key] = [val-5 for val in val_list]
```

generating X_bow, y_bow and headers:

```python
from sklearn.feature_extraction.text import CountVectorizer
# have to sort the dictionary words when giving them as we else cannot rely on index
vectorizer = CountVectorizer(vocabulary=sorted(anew_dictionary.keys()))
superhero_collection = [x[1] for x in good_heroes]+[x[1] for x in bad_heroes]

X_bow = vectorizer.fit_transform(superhero_collection)
y_bow = [0 for x in good_heroes] + [1 for x in bad_heroes]
C_bow = ["Good", "Bad"]
M_bow = sorted(anew_dictionary.keys())
```
